```python
Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
```

```python
In Bokeh, glyphs are graphical markers or symbols that you can use to display data points on a plot. Glyphs can be used to represent different types of data, such as points, lines, bars, or more complex shapes, and they play a fundamental role in creating interactive and informative data visualizations.

To add glyphs to a Bokeh plot, you typically use the figure function to create a plot, and then you add glyph renderers to that plot using methods like circle, line, or rect. These glyph renderers determine the type and appearance of the data points you want to display.

Here's an example of how to create a simple Bokeh plot with glyphs using Python:
```

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource


p = figure(width=400, height=400)


x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]


source = ColumnDataSource(data={'x': x, 'y': y})


p.circle(x='x', y='y', source=source, size=10, color='blue', legend_label='Points')

p.title.text = "Example Bokeh Plot"
p.xaxis.axis_label = "X-axis"
p.yaxis.axis_label = "Y-axis"
p.legend.title = "Legend"


show(p)
```

```python
Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?
```

```python
from bokeh.plotting import figure, show


p = figure(title="Customized Bokeh Plot", width=800, height=400, background_fill_color="lightgrey", border_fill_color="white")


x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]


source = ColumnDataSource(data={'x': x, 'y': y})


p.circle(x='x', y='y', source=source, size=10, color='blue', legend_label='Points')


p.xaxis.axis_label = "X-axis Label"
p.yaxis.axis_label = "Y-axis Label"
p.xaxis.axis_label_text_color = "blue"
p.yaxis.axis_label_text_color = "green"
p.xaxis.major_label_text_color = "red"
p.yaxis.major_label_text_color = "purple"
p.xaxis.major_tick_line_color = "orange"
p.yaxis.major_tick_line_color = "pink"
p.xaxis.minor_tick_line_color = "gray"
p.yaxis.minor_tick_line_color = "black"

p.legend.title = "Legend Title"
p.legend.label_text_font_size = "10pt"
p.legend.label_text_color = "purple"
p.legend.background_fill_color = "lightyellow"
p.legend.border_line_color = "blue"


show(p)
```

```python
Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?
```

```python

A Bokeh server is a part of the Bokeh library that allows you to create interactive web applications and dashboards with dynamic and real-time updates. With the Bokeh server, you can build web applications that include Bokeh plots and widgets, enabling users to interact with the data and see updates in real time. This is particularly useful for creating live data dashboards, scientific simulations, and other applications that require dynamic visualizations.
```

```python


from random import random

from bokeh.layouts import column
from bokeh.models import Button
from bokeh.palettes import RdYlBu3
from bokeh.plotting import figure, curdoc


p = figure(x_range=(0, 100), y_range=(0, 100), toolbar_location=None)
p.border_fill_color = 'black'
p.background_fill_color = 'black'
p.outline_line_color = None
p.grid.grid_line_color = None


r = p.text(x=[], y=[], text=[], text_color=[], text_font_size="26px",
           text_baseline="middle", text_align="center")

i = 0

ds = r.data_source


def callback():
    global i


    new_data = dict()
    new_data['x'] = ds.data['x'] + [random()*70 + 15]
    new_data['y'] = ds.data['y'] + [random()*70 + 15]
    new_data['text_color'] = ds.data['text_color'] + [RdYlBu3[i%3]]
    new_data['text'] = ds.data['text'] + [str(i)]
    ds.data = new_data

    i = i + 1


button = Button(label="Press Me")
button.on_click(callback)


curdoc().add_root(column(button, p))
```
